```python
from pathlib import Path
import pandas as pd

data_path = Path("..") / "data" / "raw" / "your_dataset.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
gender_col = "gender"
target_col = "outcome"

df.groupby(gender_col)[target_col].value_counts(normalize=True).rename("rate")
```
